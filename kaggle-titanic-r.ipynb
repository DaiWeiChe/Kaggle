```r
setwd('/kaggle/working')
x=read.csv('/kaggle/input/titanic/train.csv')
x
```

```r
summary(x)
sapply(x,function(x)
    sum(is.na(x)))
```

```r
x$Age[which(is.na(x$Age))]=29.7
summary(x)
```

```r
str(x)
x.1=x[,c(1:3,5:8,10)]
y=which(x.1$Sex=="male")
y
x.1$Sex[y]="0"
y=which(x.1$Sex=="female")
y
x.1$Sex[y]="1"
x.1
x.1$Sex=as.integer(x.1$Sex)
str(x.1)
cor(x.1)
```

```r
library(ggplot2)
library(GGally)
```

```r
ggpairs(x.1)
```

```r
model <- glm(x.1$Survived~., family = binomial(logit), x.1)
summary(model)
```

```r
model <- glm(x.1$Survived~x.1$Sex*x.1$Pclass*x.1$Age*x.1$SibSp, family = binomial(logit), x.1)
summary(model)
model
```

```r
x.test=read.csv('/kaggle/input/titanic/test.csv')
x.test
x.test$Sex[which(x.test$Sex=="male")]="0"
x.test$Sex[which(x.test$Sex=="female")]="1"
x.test$Sex=as.integer(x.test$Sex)
summary(x.test)
sapply(x.test,function(x)
    sum(is.na(x)))
x.test$Age[which(is.na(x.test$Age))]=30.27
summary(x.test)
cor(x.test[,c(1,2,4,5,6,7,9)])
```

```r
result=predict(model, newdata=x.1, type="response")
summary(result)
result_Approved <- ifelse(result > 0.55, 1, 0)
cm <- table(x.1$Survived, result_Approved, dnn = c("實際", "預測"))
cm
#(6)準確率
#計算核準卡正確率
cm[4] / sum(cm[, 2])

#計算拒補件正確率
cm[1] / sum(cm[, 1])

#整體準確率(對角線元素總和/所有觀察值總和)
accuracy <- sum(diag(cm)) / sum(cm)
accuracy
```
