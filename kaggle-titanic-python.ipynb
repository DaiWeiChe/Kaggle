## A simple guess

How about a logisitic regression model with 2 features: 
$$ y = \sigma(\alpha + \beta_1  x_1 + \beta_2  x_2 + \epsilon)$$

```python
## set up the target
y = train_df["Survived"]

## select the features
features_in_model = ["Age", "Sex"]
X = train_df[features_in_model]

## but the "Sex" is not numeric...... not allowed in a logistic regression
X
```

```python
X.Sex.replace(["male", "female"], [0, 1])
```

```python
## replace the original column with facotorized one
X["Sex"] = X.Sex.replace(["male", "female"], [0, 1])

## Still..... there are NA in "Age"
X
```

```python
## fill the NA in Age with mean value
X["Age"] = X.Age.fillna(29.7)

X
```

```python
## Scikit-learn 
from sklearn.linear_model import LogisticRegression

## set up a logistic regression model
lr_model = LogisticRegression()

## train the model with X and y
lr_model.fit(X, y)

## the evaluation method in the competition is "accuracy"
lr_model.score(X, y)
```

```python
## do all the preprocessing works on the test set

X_test = test_df[features_in_model]
X_test["Sex"] = X_test.Sex.replace(["male", "female"], [0, 1])
X_test["Age"] = X_test.Age.fillna(29.7)

## predict with test set
gender_submission["Survived"] = lr_model.predict(X_test)
gender_submission
```

## More and More

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier()

## train the model with X and y
tree_model.fit(X, y)

## the evaluation method in the competition is "accuracy"
tree_model.score(X, y)
```

```python
## predict with test set
gender_submission["Survived"] = tree_model.predict(X_test)
gender_submission

## save as csv
gender_submission.to_csv("/kaggle/working/submission.csv", index=False)
```
